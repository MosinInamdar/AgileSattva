```python
def removal_box_plot(df, column, threshold):
    sns.boxplot(df[column])
    plt.title(f'Original Box Plot of {column}')
    plt.show()

    removed_outliers = df[df[column] <= threshold]

    sns.boxplot(removed_outliers[column])
    plt.title(f'Box Plot without Outliers of {column}')
    plt.show()
    return removed_outliers


threshold_value = 8.5

no_outliers = removal_box_plot(data, 'Second year:   Sem 2', threshold_value)
print(no_outliers)
```

```python
no_outliers = removal_box_plot(data, 'Second year:   Sem 1', 8)
print(no_outliers)
```

```python
def removal_box_plot_upperlimit(df, column, threshold):
    sns.boxplot(df[column])
    plt.title(f'Original Box Plot of {column}')
    plt.show()

    removed_outliers = df[df[column] <= threshold]

    sns.boxplot(removed_outliers[column])
    plt.title(f'Box Plot without Outliers of {column}')
    plt.show()
    return removed_outliers


no_outliers = removal_box_plot_upperlimit(data, 'First year:   Sem 1', 20)
no_outliers
```

```python
no_outliers = removal_box_plot_upperlimit(data, 'First year:   Sem 2', 20)
no_outliers
```

```python

def remove_outlier(df, columns):
  for col in columns:
    # Calculate the interquartile range (IQR)
    IQR = np.percentile(df[col], 75) - np.percentile(df[col], 25)

    # Calculate the upper and lower bounds
    upper_bound = np.percentile(df[col], 75) + 1.5 * IQR
    lower_bound = np.percentile(df[col], 25) - 1.5 * IQR

    # Replace outliers with the bounds
    df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])
    df[col] = np.where(df[col] < lower_bound, lower_bound, df[col])

  return df

# Apply the function to the studentData dataframe
numeric_columns = data.select_dtypes(include=[np.number]).columns
data = remove_outlier(data, numeric_columns)

# Check for outliers again
data.describe()
```
